# 多元逻辑回归

逻辑回归就是在线性回归的计算结果上，增加了一个Sigmoid激活函数。
$$
LogisticRegression(x) = Sigmoid(LinearRegression(x)) = Sigmoid(wx+b)
$$

Sigmoid这个激活函数的作用可以将大于0的数快速映射到接近1，小于0的数快速映射到接近0。



## 多元逻辑回归
$$
LogisticRegression(x_1,x_2,...,x_n)=Sigmoid(LinearRegression(x_1,x_2,...,x_n))

=Sigmond(w_1x_1+w_2x_2+...w_nx_n+b)

### 逻辑回归的决策边界
输入特征有2个，分别用x轴和y轴表示，进行二分类。但是它的决策边界不是直线。

——> 构造高次项特征让线性回归拟合曲线！！

$$Sigmoid(w_1x_1^2+w_2x_2^2+b)$$
$ w_1x_1^2+w_2x_2^2+b=0 $ 就是决策边界的曲线方程



如果你增加更多的高次项，理论上逻辑回归的决策边界可以是任意曲线。
